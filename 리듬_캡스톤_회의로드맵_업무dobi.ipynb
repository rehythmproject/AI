```python
import openai

def generate_feedback_from_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        meeting_notes = file.read()

    messages = [
        {"role": "system", "content": "기술개발등을 할 때 그에 맞는 로드맵과 계획 날짜 등을 정리해 줘."},
        {"role": "user", "content": f"회의 내용: {meeting_notes}"}
    ]

    response = openai.ChatCompletion.create(
        model="gpt-4o",
        messages=messages,
        max_tokens=650,
        temperature=0.5,
        stop=None
    )

    feedback = response['choices'][0]['message']['content']
    return feedback

file_path = '/content/김유진 오늘 모인 주 목적은 신규 웹 서비스 개발을 위한 기술.txt'

feedback = generate_feedback_from_file(file_path)
print("업무 dobi:")
print(feedback)
```
