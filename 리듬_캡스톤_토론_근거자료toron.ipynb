```python
import openai

def generate_feedback_from_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        meeting_notes = file.read()

    messages = [
        {"role": "system", "content": "회사내에서 토론이나 개인적인 토론시 그의 기반되는 근거자료들을 찾아줘."},
        {"role": "user", "content": f"회의 내용: {meeting_notes}"}
    ]
    response = openai.ChatCompletion.create(
        model="gpt-4o",
        messages=messages,
        max_tokens=700,
        temperature=0.5,
        stop=None
    )

    feedback = response['choices'][0]['message']['content']
    return feedback

# 파일 경로 설정
file_path = '/content/진행자 여러분, 오늘 우리는 기후 변화와 그 영향을 주제로 토론.txt'

feedback = generate_feedback_from_file(file_path)
print("회의 toron:")
print(feedback)
```
